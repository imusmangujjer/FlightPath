## **2.b.Standardize Datatypes:**

**of some of object into date & time in order to:**

**Accurate Calculations:** Time calculations (e.g., differences, durations) require datetime objects.

**Enhanced Analysis:** Enables grouping, sorting, and filtering by time/date.

**Visualization:** Time-series analysis requires proper datetime types.


**Also we change the datatype of "flight number" object cause it is a descriptive feature rather than numeric**


```python
df_typed=df_cleaned.copy()

# Convert "Date (MM/DD/YYYY)" column to datetime, coercing errors
df_typed["Date (MM/DD/YYYY)"] = pd.to_datetime(df_typed["Date (MM/DD/YYYY)"], errors='coerce')

# If you need to display the date in a specific format elsewhere:
# df_typed["Formatted Date"] = df_typed["Date (MM/DD/YYYY)"].dt.strftime('%m/%d/%y')

# Ensure "Flight Number" is string
df_typed["Flight Number"] = df_typed["Flight Number"].astype(str)

# Ensure all time columns are processed correctly
time_columns = [
    "Scheduled departure time",
    "Actual departure time",
    "Wheels-off time",
    "Scheduled Arrival Time",
    "Actual Arrival Time",
    "Wheels-on Time"
]

# Define a function to normalize and convert time formats
def normalize_time_format(time_value):
    try:
        # Convert to string and split the time into parts
        parts = str(time_value).split(':')

        # Handle cases like h:m, h:mm, hh:m, hh:mm
        if len(parts) == 2 and parts[0].isdigit() and parts[1].isdigit():
            hours = parts[0].zfill(2)  # Ensure 2 digits for hours
            minutes = parts[1].zfill(2)  # Ensure 2 digits for minutes
            return f"{hours}:{minutes}"
    except Exception:
        pass  # Ignore invalid values
    return '00:00'  # Default for invalid or missing values

# Process each time column
for column in time_columns:
    # Replace missing values with '00:00' and normalize the format
    df_typed[column] = df_typed[column].fillna('00:00').astype(str).apply(normalize_time_format)

    # Convert to datetime, coercing invalid formats to 'NaT'
    df_typed[column] = pd.to_datetime(df_typed[column], format='%H:%M', errors='coerce')

df_typed.info()
```

```python
import gradio as gr
import pandas as pd
from pycaret.regression import load_model, predict_model

# Load the best trained model
best_model = load_model("best_regression_model")

# Function to format the prediction output
def format_prediction(value):
    minutes = int(value)
    seconds = int((value - minutes) * 60)
    return f"{minutes}Min {seconds}s"

# Function to make predictions
def predict_delay(day, month, year, taxi_out, taxi_in, origin, carrier, sched_time, actual_time, sched_hour):
    try:
        input_data = pd.DataFrame({
            "Day": [day], "Month": [month], "Year": [year],
            "Taxi-Out time (Minutes)": [taxi_out], "Taxi-In time (Minutes)": [taxi_in],
            "Origin Airport": [origin],
            "Carrier Code": [carrier], "Scheduled Elapsed Time (Minutes)": [sched_time],
            "Actual Elapsed Time (Minutes)": [actual_time], "Scheduled departure hour": [sched_hour]
        })

        prediction = predict_model(best_model, data=input_data)

        predicted_column = prediction.columns[-1]
        prediction.rename(columns={predicted_column: "Predictive Arrival Delay (Minutes)"}, inplace=True)
        formatted_prediction = format_prediction(prediction["Predictive Arrival Delay (Minutes)"].values[0])

        return prediction, formatted_prediction
    except Exception as e:
        return f"Error: {str(e)}", None

# Gradio UI
description = """### Flight Arrival Delay Prediction System\n\nEnter flight details to predict the estimated arrival delay time."""

with gr.Blocks() as demo:
    gr.Markdown(description)

    # Input fields
    with gr.Row():
        day = gr.Number(label="Day", value=1, minimum=1, maximum=31)
        month = gr.Number(label="Month", value=2, minimum=1, maximum=12)
        year = gr.Number(label="Year", value=2025)

    with gr.Row():
        taxi_out = gr.Number(label="Taxi-Out Time (Minutes)", value=10)
        taxi_in = gr.Number(label="Taxi-In Time (Minutes)", value=5)

    with gr.Row():
        origin = gr.Dropdown(label="Origin Airport", choices=["ATL", "DFW", "DEN", "LAX", "ORD"], value="ATL")
        carrier = gr.Dropdown(label="Carrier Code", choices=["AA", "DL"], value="AA")

    with gr.Row():
        sched_time = gr.Number(label="Scheduled Elapsed Time (Minutes)", value=120)
        actual_time = gr.Number(label="Actual Elapsed Time (Minutes)", value=130)
        sched_hour = gr.Number(label="Scheduled Departure Hour", value=14, minimum=0, maximum=23)

    # Predict button and result
    predict_btn = gr.Button("Predict Arrival Delay")
    result = gr.Textbox(label="Predictive Arrival Delay (Minutes)")
    dataframe_output = gr.Dataframe(label="Prediction Details")

    def predict_and_display(*args):
        df, prediction = predict_delay(*args)
        return prediction, df

    predict_btn.click(predict_and_display, inputs=[day, month, year, taxi_out, taxi_in, origin, carrier, sched_time, actual_time, sched_hour], outputs=[result, dataframe_output])

# Launch the Gradio app
demo.launch(show_error=False, debug=False)
```
