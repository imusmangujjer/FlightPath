```python
df_typed=df_cleaned.copy()

# Convert "Date (MM/DD/YYYY)" column to datetime, coercing errors
df_typed["Date (MM/DD/YYYY)"] = pd.to_datetime(df_typed["Date (MM/DD/YYYY)"], errors='coerce')

# If you need to display the date in a specific format elsewhere:
# df_typed["Formatted Date"] = df_typed["Date (MM/DD/YYYY)"].dt.strftime('%m/%d/%y')

# Ensure "Flight Number" is string
df_typed["Flight Number"] = df_typed["Flight Number"].astype(str)

# Ensure all time columns are processed correctly
time_columns = [
    "Scheduled departure time",
    "Actual departure time",
    "Wheels-off time",
    "Scheduled Arrival Time",
    "Actual Arrival Time",
    "Wheels-on Time"
]

# Define a function to normalize and convert time formats
def normalize_time_format(time_value):
    try:
        # Convert to string and split the time into parts
        parts = str(time_value).split(':')

        # Handle cases like h:m, h:mm, hh:m, hh:mm
        if len(parts) == 2 and parts[0].isdigit() and parts[1].isdigit():
            hours = parts[0].zfill(2)  # Ensure 2 digits for hours
            minutes = parts[1].zfill(2)  # Ensure 2 digits for minutes
            return f"{hours}:{minutes}"
    except Exception:
        pass  # Ignore invalid values
    return '00:00'  # Default for invalid or missing values

# Process each time column
for column in time_columns:
    # Replace missing values with '00:00' and normalize the format
    df_typed[column] = df_typed[column].fillna('00:00').astype(str).apply(normalize_time_format)

    # Convert to datetime, coercing invalid formats to 'NaT'
    df_typed[column] = pd.to_datetime(df_typed[column], format='%H:%M', errors='coerce')

df_typed.info()
```

```python
# calculating percentage of outliers in the dataset

# Define a function to calculate the whiskers
def whisker_range_function(d):
    outlier_list = []
    Numeric_variable_list = []
    lower_whisker = []
    upper_whisker = []

    for i in d.columns:
        # Check for numerical variables only
        if pd.api.types.is_numeric_dtype(d[i]):
            # Remove NaN values to avoid errors
            clean_data = d[i].dropna()

            # Calculate quartiles and IQR
            Q1 = np.percentile(clean_data, 25)
            Q3 = np.percentile(clean_data, 75)
            IQR = Q3 - Q1
            Numeric_variable_list.append(i)

            # Calculate whisker range
            whisker_range = 1.5 * IQR
            whisker_l = Q1 - whisker_range
            whisker_u = Q3 + whisker_range

            # Count outliers and append to the list
            lower_outliers = len(clean_data[clean_data < whisker_l])
            upper_outliers = len(clean_data[clean_data > whisker_u])

            lower_whisker.append(f'{whisker_l:0.1f}')
            upper_whisker.append(f'{whisker_u:0.1f}')
            outlier_list.append((lower_outliers, upper_outliers))

    return Numeric_variable_list, lower_whisker, upper_whisker, outlier_list

# Call the function
result = whisker_range_function(df_typed)

# Unpack the results
Numeric_variable_list, lower_whisker, upper_whisker, outlier_list = result

# Create a DataFrame
outlier_df = pd.DataFrame({
    'Lower_Whisker': lower_whisker,
    'Upper_Whisker': upper_whisker,
    'Negative_Outliers': [t[0] for t in outlier_list],
    'Positive_Outliers': [t[1] for t in outlier_list],
    'Percentage_Negative_Outliers': [round((t[0] / len(df_typed)) * 100, 3) for t in outlier_list],
    'Percentage_Positive_Outliers': [round((t[1] / len(df_typed)) * 100, 3) for t in outlier_list]
}, index=Numeric_variable_list)

# Display the DataFrame
outlier_df
```

# **5. Hypothesis Testing:**

## Do flights departing from certain airports experience longer delays?

```python
from scipy.stats import f_oneway

# Group data by 'Origin Airport' and collect arrival delays
grouped_arrival_delays = [df_new[df_new['Origin Airport'] == airport]['Arrival Delay (Minutes)'].dropna()
                          for airport in df_new['Origin Airport'].unique()]

# Perform one-way ANOVA
anova_result_arrival = f_oneway(*grouped_arrival_delays)

# Display the results
print("ANOVA Test Results for Arrival Delays:")
print(f"F-statistic: {anova_result_arrival.statistic}")
print(f"P-value: {anova_result_arrival.pvalue}")

# Interpretation
if anova_result_arrival.pvalue < 0.05:
    print("Reject the null hypothesis: At least one origin airport has a significantly different arrival delay.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in arrival delays among origin airports.")
```

F-Statistic:

633.36 This is a very high F-statistic, indicating that the variation in mean Arrival Delays between different origin airports is much larger than the variation within the individual airport groups. In simple terms, the average arrival delays for flights differ significantly depending on the airport of departure.

P-value:

0.0 This is an extremely small p-value (almost zero), far below the significance level of 0.05. A p-value this small means that the observed differences between the mean arrival delays of origin airports are not due to random chance.

Conclusion:

Reject the Null Hypothesis. The null hypothesis (that all origin airports have the same average arrival delays) is rejected. The alternative hypothesis is accepted: There are significant differences in arrival delays among origin airports.

Key Insight:

The airport you depart from does matter for how delayed you are when you arrive. Some airports cause longer arrival delays than others.

Possible Causes:

Congestion at certain origin airports could lead to downstream arrival delays. Operational inefficiencies or weather conditions at specific airports might contribute to longer delays. Flights from some airports might face more frequent air traffic control delays or routing issues.

### **Tukey’s Honest Significant Difference (HSD) test**

```python
from scipy.stats import f_oneway
import pandas as pd
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Define hypotheses
print("\nOne-Way ANOVA Test: Origin Airport vs. Arrival Delay")
print("Null Hypothesis (H₀): There is no significant difference in arrival delays among origin airports.")
print("Alternative Hypothesis (H₁): At least one origin airport has a significantly different arrival delay.\n")

# Group data by 'Origin Airport' and collect arrival delays
grouped_arrival_delays = [df_new[df_new['Origin Airport'] == airport]['Arrival Delay (Minutes)'].dropna()
                          for airport in df_new['Origin Airport'].unique()]

# Perform one-way ANOVA
anova_result_arrival = f_oneway(*grouped_arrival_delays)

# Display ANOVA results
print("ANOVA Test Results for Arrival Delays:")
print(f"F-statistic: {anova_result_arrival.statistic:.4f}")
print(f"P-value: {anova_result_arrival.pvalue:.4e}")

# Interpretation
alpha = 0.05
if anova_result_arrival.pvalue < alpha:
    print("\nConclusion: Reject the null hypothesis (H₀).")
    print("➡ There are significant differences in arrival delays among origin airports.")

    # Perform Tukey HSD post-hoc test
    print("\nPerforming Tukey HSD Test to identify which airports differ...\n")

    tukey_result = pairwise_tukeyhsd(df_new['Arrival Delay (Minutes)'], df_new['Origin Airport'], alpha=0.05)
    print(tukey_result)

else:
    print("\nConclusion: Fail to reject the null hypothesis (H₀).")
    print("➡ No significant differences in arrival delays among origin airports.")
```

## Does the airline affect the average delay time of flights? (TNV)

```python
# Null Hypothesis (H₀): There are no significant differences in average delay times among the carriers (Any observed differences are likely due to random chance).
# Alternative Hypothesis (H₁): There are significant differences in average delay times among the carriers (Some carriers consistently experience longer or shorter delays than others)

from scipy.stats import ttest_ind

# Filter data for two specific carriers
carrier_aa_delays = df_new[df_new['Carrier Code'] == 'AA']['Arrival Delay (Minutes)'].dropna()
carrier_dl_delays = df_new[df_new['Carrier Code'] == 'DL']['Arrival Delay (Minutes)'].dropna()

# Perform an independent t-test
t_test_result = ttest_ind(carrier_aa_delays, carrier_dl_delays, equal_var=False)
# Display the results
print("T-Test Results for Delay Times Between Two Carriers:")
print(f"T-statistic: {t_test_result.statistic}")
print(f"P-value: {t_test_result.pvalue}")

# Interpretation
if t_test_result.pvalue < 0.05:
    print("Reject the null hypothesis: There is a significant difference in average delay times between the two carriers.")
else:
    print("Fail to reject the null hypothesis: No significant difference in average delay times between the two carriers.")
```

This situation arises due to statistical vs. practical significance:

Statistical Significance

Hypothesis tests (e.g., t-tests, ANOVA) determine whether the observed difference in means is unlikely due to random chance.
If your sample size is very large, even tiny differences (like 0.25 seconds) can result in a very small p-value, leading to statistical significance.
Practical Significance

Even though the difference is statistically significant, a 0.25-second delay is not meaningful in real-world terms.
Airlines, passengers, and operations likely wouldn’t care about such a small variation in delay.

**Hypothesis and Motivation**
The hypothesis aimed to determine whether the airline affects the average delay time of flights, focusing on arrival delays as the target variable. Arrival delays are crucial because they directly impact customer satisfaction, operational efficiency, and airline reputation. Identifying significant differences in delay performance allows airlines to address areas that contribute to poor outcomes.


Null Hypothesis (H₀): There are no significant differences in average delay times between the two carriers.

Alternative Hypothesis (H₁): There are significant differences in average delay times between the two carriers.

The t-test was selected as the method for comparing the mean arrival delays of the two carriers, AA and DL.

**Why Use a T-Test?**
The independent t-test was chosen for the following reasons:

Comparison of Means:
The hypothesis centers on comparing the average (mean) arrival delays between the two airlines. The t-test is specifically designed to test whether the means of two independent groups are statistically different.

Two Independent Groups:
The data involves two distinct groups (AA and DL), with delay times for one airline being independent of the other.

Continuous Target Variable:
Arrival delays are continuous numerical values, which is a prerequisite for using the t-test.

Robustness of the T-Test:
The t-test is robust to minor deviations from normality, making it suitable for real-world data like flight delays, especially with reasonably large sample sizes.

**T-Statistic:** -9.47
This indicates a significant difference in the average delay times between the two carriers. The negative sign shows which carrier has shorter delays on average. The magnitude of this value reflects a large difference relative to the variability in the data.

**P-Value:** 2.94 × 10⁻²¹
The p-value is astronomically small, much lower than the common significance threshold of 0.05. This confirms that the difference in delay times is not due to random chance.

**Conclusion**
The results of the Welch’s t-test show a significant difference in average delay times between the two carriers. This suggests that operational or systemic factors may be affecting performance differently for each airline. These findings provide a strong basis for advising the airlines to analyze and address the specific areas contributing to delays, thereby improving customer satisfaction and operational efficiency.

## There is no difference in flight delays between winter months (Dec–Feb) and summer months (Jun–Aug). -EMMA


𝐻 0 : 𝜇 Winter = 𝜇 Summer

𝐻 1 : 𝜇 Winter ≠ 𝜇 Summer


```python
from scipy.stats import ttest_ind

# Convert 'Date (MM/DD/YYYY)' to datetime objects (if not already done)
df_new['Date (MM/DD/YYYY)'] = pd.to_datetime(df_new['Date (MM/DD/YYYY)'])

# Create separate DataFrames for winter and summer months
winter_months = df_new[(df_new['Date (MM/DD/YYYY)'].dt.month >= 12) | (df_new['Date (MM/DD/YYYY)'].dt.month <= 2)]
summer_months = df_new[(df_new['Date (MM/DD/YYYY)'].dt.month >= 6) & (df_new['Date (MM/DD/YYYY)'].dt.month <= 8)]

# Extract Arrival Delay (Minutes) data for winter and summer months
winter_delays = winter_months['Arrival Delay (Minutes)'].dropna()
summer_delays = summer_months['Arrival Delay (Minutes)'].dropna()

# Perform an independent samples t-test
t_statistic, p_value = ttest_ind(winter_delays, summer_delays, equal_var=False)

# Print the results
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in delays between winter and summer months.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in delays between winter and summer months.")
```

* T-Statistic: -25.0205147777957

* P-Value: 7.426016352714923e-138(extremely small)
A very low p-value (< 0.05) indicates that the observed difference in delays is statistically significant.

* Conclusion:
Reject the Null Hypothesis: There is strong evidence that flight delays during winter months are different from (and likely more severe than) those in summer months.

```python
import pandas as pd
from scipy.stats import mannwhitneyu

# Hypotheses:
# H0: The distribution of arrival delays before 2020 is the same as after 2020.
# H1: The distribution of arrival delays before 2020 is different from after 2020.

# Filter data based on the year
df_new['Year'] = pd.to_datetime(df_new['Date (MM/DD/YYYY)']).dt.year
before_2020 = df_new[df_new['Year'] < 2020]['Arrival Delay (Minutes)'].dropna()
after_2020 = df_new[df_new['Year'] > 2020]['Arrival Delay (Minutes)'].dropna()

# Perform Mann-Whitney U test
u_stat, p_value = mannwhitneyu(before_2020, after_2020, alternative='two-sided')

# Print Mann-Whitney results
print("Mann-Whitney U Test")
print("-------------------")
print(f"Hypotheses: \nH0: The distribution of arrival delays before 2020 is the same as after 2020. \nH1: The distribution of arrival delays is different.")
print(f"Mann-Whitney U Statistic: {u_stat}")
print(f"P-value: {p_value}")

# Conclusion
alpha = 0.05
if p_value < alpha:
    print(f"Conclusion: Reject the null hypothesis (H0). There is a significant difference in the arrival delay distributions before and after 2020 (excluding 2020).")
else:
    print(f"Conclusion: Fail to reject the null hypothesis (H0). There is no significant difference in the arrival delay distributions before and after 2020 (excluding 2020).")
```

```python
import pandas as pd
import numpy as np

# Hypotheses:
# H0: There is no difference in the mean arrival delays before and after 2020.
# H1: There is a difference in the mean arrival delays before and after 2020.

# Filter data based on the year
df_new['Year'] = pd.to_datetime(df_new['Date (MM/DD/YYYY)']).dt.year
before_2020 = df_new[df_new['Year'] < 2020]['Arrival Delay (Minutes)'].dropna()
after_2020 = df_new[df_new['Year'] > 2020]['Arrival Delay (Minutes)'].dropna()

# Bootstrap analysis
def bootstrap_ci(data1, data2, n_bootstrap=100, ci=95):
    np.random.seed(42)
    boot_differences = []
    for _ in range(n_bootstrap):
        sample1 = np.random.choice(data1, size=len(data1), replace=True)
        sample2 = np.random.choice(data2, size=len(data2), replace=True)
        boot_differences.append(np.mean(sample2) - np.mean(sample1))
    lower = np.percentile(boot_differences, (100 - ci) / 2)
    upper = np.percentile(boot_differences, 100 - (100 - ci) / 2)
    return lower, upper

# Calculate bootstrap confidence interval
ci_lower, ci_upper = bootstrap_ci(before_2020.values, after_2020.values)

# Print Bootstrap results
print("\nBootstrap Analysis")
print("-------------------")
print(f"Hypotheses: \nH0: There is no difference in the mean arrival delays before and after 2020. \nH1: There is a difference in the mean arrival delays.")
print(f"Bootstrap Mean Difference CI (95%): ({ci_lower:.2f}, {ci_upper:.2f})")

# Conclusion
if ci_lower > 0 or ci_upper < 0:
    print(f"Conclusion: Reject the null hypothesis (H0). There is a significant difference in the mean arrival delays before and after 2020 (excluding 2020).")
else:
    print(f"Conclusion: Fail to reject the null hypothesis (H0). There is no significant difference in the mean arrival delays before and after 2020 (excluding 2020).")
```
