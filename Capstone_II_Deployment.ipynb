```python
import gradio as gr
import pandas as pd
import time
import tensorflow as tf
import joblib
import xgboost as xgb
import openai
import numpy as np

# ✅ Load MLP Neural Network Model
mlp_model = tf.keras.models.load_model("/content/drive/MyDrive/capstone1/All_Airlines/mlp_model.keras")

# ✅ Load Scaler & Encoder for Preprocessing
scaler = joblib.load("/content/drive/MyDrive/capstone1/All_Airlines/scaler.pkl")
encoder = joblib.load("/content/drive/MyDrive/capstone1/All_Airlines/encoder.pkl")

# ✅ Load the combined airline reviews dataset
file_path = "/content/drive/MyDrive/capstone1/Sentiment Analysis/Reviews.csv"
df_reviews = pd.read_csv(file_path)

# ✅ OpenAI API Key (Replace with your actual API key)
openai.api_key = "Your-Open AI-API-Key"

# ✅ Function to extract most positive and most negative reviews
def get_reviews(carrier_code):
    df_carrier = df_reviews[df_reviews["Carrier Code"] == carrier_code]

    if df_carrier.empty:
        return "No reviews available for this carrier.", "No reviews available for this carrier."

    most_positive = df_carrier.loc[df_carrier["Sentiment"].idxmax(), "Review"]
    most_negative = df_carrier.loc[df_carrier["Sentiment"].idxmin(), "Review"]

    return most_positive, most_negative

# ✅ Function to format the prediction output
def format_prediction(value):
    minutes = int(value)
    seconds = int((value - minutes) * 60)
    return f"{minutes} Min {seconds} s"

# ✅ Function to generate AI-based recommendations using OpenAI GPT-4
def generate_gpt4_recommendations(carrier, origin, destination, delay, taxi_in, taxi_out, sched_time, dep_hour, dep_delay):
    prompt = f"""
    You are an airline operations expert. Given the following flight details, generate a practical recommendation to minimize delays, Note: if the delay is in negative you can still suggest some improvements but
    you can also praise the efficiency of the airline:

    - Carrier: {carrier}
    - Origin Airport: {origin}
    - Destination Airport: {destination}
    - Predicted Arrival Delay: {delay:.2f} minutes
    - Taxi-In Time: {taxi_in} minutes
    - Taxi-Out Time: {taxi_out} minutes
    - Scheduled Elapsed Time: {sched_time} minutes
    - Departure Hour: {dep_hour}:00
    - Departure Delay: {dep_delay} minutes

    🎯 Provide 3 actionable suggestions that could help the airline improve efficiency.
    """

    try:
        response = openai.chat.completions.create(
            model="gpt-4",
            messages=[{"role": "system", "content": prompt}],
        )
        return response.choices[0].message.content

    except Exception as e:
        return f"⚠️ AI Recommendation Error: {str(e)}"

# ✅ Function to predict using XGBoost & MLP, and get the average
def predict_delay(day, month, year, taxi_out, taxi_in, origin, carrier, dest, sched_time, sched_hour, dep_delay):
    try:
        if origin == dest:
            return "❌ Error: Origin and Destination cannot be the same.", None, None, "", "", ""

        # Simulate a loading delay for better UX
        time.sleep(2)

        # Prepare categorical inputs
        categorical_data = np.array([[origin, carrier, dest]])
        encoded_categorical = encoder.transform(categorical_data)

        # Prepare numerical inputs
        numerical_data = np.array([[day, month, year, taxi_out, taxi_in, sched_time, sched_hour, dep_delay]])
        scaled_numerical = scaler.transform(numerical_data)

        # Combine all features
        X_input = np.hstack((scaled_numerical, encoded_categorical))

        # ✅ Predict using MLP (Neural Network)
        mlp_prediction = mlp_model.predict(X_input)[0][0]
        final_prediction = mlp_prediction
        formatted_prediction = format_prediction(final_prediction)

        # ✅ Confidence Interval (based on test set analysis)
        mae = 5.17  # Mean Absolute Error from evaluation
        residual_std = 7.23  # Residual standard deviation from test set
        ci_range = 1.96 * residual_std
        upper_bound = final_prediction + ci_range
        lower_bound = final_prediction - ci_range

        # ✅ Friendly explanation string
        explanation = (
            f"{formatted_prediction} ⏱️\n\n"
            f"✈️ Based on similar past flights, this prediction is expected to be accurate "
            f"within **±{ci_range:.2f} minutes** most of the time (95% confidence).\n\n"
            f"💡 On average, our model’s predictions are off by about **±{mae:.2f} minutes** compared to actual delays.\n\n"
            f"That means this flight could actually arrive anywhere from **~{int(lower_bound)} to ~{int(upper_bound)} minutes delayed**, "
            f"depending on real-world factors.\n\n"
            f"We used deep learning model trained on real flight data to provide the most accurate estimate possible."
        )

        # ✅ Get airline-specific reviews
        best_review, worst_review = get_reviews(carrier)

        # ✅ Generate AI-based recommendation using GPT-4
        ai_recommendation = generate_gpt4_recommendations(
            carrier, origin, dest, final_prediction,
            taxi_in, taxi_out, sched_time, sched_hour, dep_delay
        )

        return explanation, best_review, worst_review, ai_recommendation

    except Exception as e:
        return f"❌ Error: {str(e)}", None, None, ""

# ✅ Gradio UI
description = """
# ✈️ Flight Arrival Delay Prediction System
🔹 Enter flight details to **predict the estimated arrival delay time**.

🔹 Ensure **Origin & Destination are different**.
"""

def validate_airports(origin, dest):
    if origin == dest:
        return "❌ Origin and Destination cannot be the same. Please select different airports."
    return None

with gr.Blocks() as demo:
    with gr.Row():
        gr.Image("/content/drive/MyDrive/capstone1/All_Airlines/American Airlines.png", elem_id="logo-left", width=100, height=100)
        gr.Markdown(description, elem_id="title", elem_classes=["center"])
        gr.Image("/content/drive/MyDrive/capstone1/All_Airlines/Delta.png", elem_id="logo-right", width=100, height=100)

    gr.Markdown("---")

    gr.Markdown("## 📆 Flight Information", elem_classes=["center"])
    with gr.Row():
        day = gr.Number(label="Day", value=1, minimum=1, maximum=31)
        month = gr.Number(label="Month", value=2, minimum=1, maximum=12)
        year = gr.Number(label="Year", value=2025)

    gr.Markdown("---")

    gr.Markdown("## ⏳ Delay & Time Information", elem_classes=["center"])
    with gr.Row():
        taxi_out = gr.Number(label="Taxi-Out Time (Minutes)", value=10)
        taxi_in = gr.Number(label="Taxi-In Time (Minutes)", value=5)

    with gr.Row():
        sched_time = gr.Number(label="Scheduled Elapsed Time (Minutes)", value=130)
        sched_hour = gr.Number(label="Scheduled Departure Hour", value=14, minimum=0, maximum=23)
        dep_delay = gr.Number(label="Departure Delay (Minutes)", value=15)

    gr.Markdown("---")

    gr.Markdown("## 🌍 Flight Route Information", elem_classes=["center"])
    with gr.Row():
        origin = gr.Dropdown(label="Origin Airport", choices=["ATL", "DFW", "DEN", "LAX", "ORD"], value="ATL")
        carrier = gr.Dropdown(label="Carrier Code", choices=["AA", "DL"], value="AA")
        dest = gr.Dropdown(label="Destination Airport", choices=["ATL", "DFW", "DEN", "LAX", "ORD"], value="DFW")

    gr.Markdown("---")

    predict_btn = gr.Button("🚀 Predict Arrival Delay")
    result = gr.Textbox(label="🔮 Predictive Arrival Delay (Minutes)")
    best_comment = gr.Textbox(label="👍 Best Comment")
    worst_comment = gr.Textbox(label="👎 Worst Comment")
    recommendation = gr.Textbox(label="🤖 AI-Generated Recommendation")

    predict_btn.click(predict_delay, inputs=[day, month, year, taxi_out, taxi_in, origin, carrier, dest, sched_time, sched_hour, dep_delay], outputs=[result, best_comment, worst_comment, recommendation])

demo.launch(pwa=True)
```

```python
import numpy as np
import joblib

# Load scaler, encoder, and the trained XGBoost delay model
scaler = joblib.load("scaler.pkl")
encoder = joblib.load("encoder.pkl")
xgb_model = joblib.load("xgboost_model.pkl")

# Input values (from UI / screenshot)
day = 1
month = 2
year = 2025
taxi_out = 10
taxi_in = 5
sched_time = 130
sched_hour = 14
dep_delay = 15
origin = "ATL"
carrier = "AA"
dest = "LAX"

# Prepare input data
numerical_data = np.array([[day, month, year, taxi_out, taxi_in,
                            sched_time, sched_hour, dep_delay]])
scaled_numerical = scaler.transform(numerical_data)

categorical_data = np.array([[origin, carrier, dest]])
encoded_categorical = encoder.transform(categorical_data)

# Final input
X_input = np.hstack((scaled_numerical, encoded_categorical))

# Predict
predicted_delay = xgb_model.predict(X_input)[0]
print(f"📦 Predicted Arrival Delay: {predicted_delay:.2f} minutes")
```

```python
from pycaret.regression import load_model, predict_model
import pandas as pd

# Load PyCaret model
model = load_model("best_regression_model")

# Input data
new_data = pd.DataFrame({
    "Day": [1],
    "Month": [2],
    "Year": [2025],
    "Taxi-Out time (Minutes)": [10],
    "Taxi-In time (Minutes)": [5],
    "Scheduled Elapsed Time (Minutes)": [130],
    "Scheduled departure hour": [14],
    "Departure delay (Minutes)": [15],
    "Origin Airport": ["ATL"],
    "Carrier Code": ["AA"],
    "Destination Airport": [""]
})

# Predict
prediction = predict_model(model, data=new_data)

# Inspect the full prediction output
print("\n🔍 Prediction Output:\n", prediction)

# Access the predicted value safely (check which column holds it)
predicted_col = prediction.columns[-1]  # last column is usually the prediction
predicted_delay = prediction[predicted_col].iloc[0]

print(f"\n📦 Predicted Arrival Delay: {predicted_delay:.2f} minutes")
```
